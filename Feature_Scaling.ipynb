```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df = pd.read_csv('data/data_cleaned.csv')
```

```python
df.head()
```

```python
numerical_features = ['Height (cm)', 'Width (cm)', 'Depth (cm)', 'Avg Daily Demand']
X = df[numerical_features]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df_scaled = df.copy()
df_scaled[numerical_features] = X_scaled
```

```python
df_scaled.to_csv('data/feature_scaled.csv', index=False)
```
